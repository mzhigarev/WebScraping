# ScrapingTutorial

## importing libs and url

```python
import bs4 as bs
import urllib.request
import http.cookiejar
```

```python
url = "http://play.google.com/store/apps/details?id=air.air.cookingstory"
```

```python
sauce = urllib.request.urlopen(url).read()
```

```python
soup = bs.BeautifulSoup(sauce,"lxml")
```

## trying BS

```python
#getting the title
print(soup.title.text)
```

```python
#print link text
for link_text in soup.find_all('a'):
    print(link_text.string)
```

```python
#print link url
for url in soup.find_all('a'):
    print(url.get('href'))
```

```python
for div in soup.find_all('div', class_='meta-info'):
    print(div.contents[1].text.strip()," --- ",div.contents[3].text.strip())
```

## lenta.ru

```python
#getting links for navigation bar
nav = soup.nav
for url in nav.find_all('a'):
    print(url.get('href'))
```

```python
#getting article names
for div in soup.find_all('div', class_="item article"):
    print(div.text,"\n")
```
